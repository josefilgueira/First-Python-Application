```python
import getnumber as gn #Importa librería local con funciones para pedir y comprobar números. (opción de pedir número oculto)
import random
import matplotlib.pyplot as grafico
from faker import Faker #Debemos instalar esta librería para que funcione (pip install faker). Nos ayudará a crear registros ficticios.
from openpyxl import Workbook, load_workbook

# Al principio creo el código de las funciones para luego aplicarlas al juego.

#Función que abre las estadísticas guardadas. En caso de no existir el archivo, lo crea.
def abrir_estad():
    try:
        wb = load_workbook('estad_tareaJF.xlsx')
    except FileNotFoundError: #En este caso no existe el archivo, entonces lo crea.
        wb = Workbook()
        ws = wb.active
        ws.title = 'Estadísticas'
        encabezados = ['Jugador','Modo de Juego','Dificultad Intentos','Dificultad Rangos', 'Partidas Jugadas', 'Partidas Ganadas','Partidas Perdidas', 
                       'Intentos Totales', 'Promedio de Intentos','Porcentaje de éxito', ]
        ws.append(encabezados)
        wb.save('estad_tareaJF.xlsx')
    return wb

#Actualiza las estadísticas filtrando según usuario,  modo de juego y dificultad. Si encuentra un registro lo actualiza y sino lo crea.
def actualizar_estad(Jugador, Modo, Dificultad1, Dificultad2, Resultado, Intentos):
    wb = load_workbook('estad_tareaJF.xlsx')
    ws = wb['Estadísticas']

    #Busca filtrando por jugador modo y diferentes dificultades
    encontrado = False
    for row in range(2, ws.max_row + 1):#Recorre todas las filas sin tener en cuenta el encabezado
        if (ws[f'A{row}'].value == Jugador and ws[f'B{row}'].value == Modo  and ws[f'C{row}'].value == Dificultad1 and ws[f'D{row}'].value == Dificultad2) :
            #Al encontrar una fila que coincida con los filtros indicados, procede a su actualización y acaba el for.
            encontrado = True
            partidas_jugadas = ws[f"E{row}"].value + 1
            partidas_ganadas = ws[f"F{row}"].value + (1 if Resultado else 0)
            partidas_perdidas = ws[f"G{row}"].value + (0 if Resultado else 1)
            intentos_totales = ws[f"H{row}"].value + Intentos
            promedio_intentos = intentos_totales / partidas_jugadas
            porcentaje_exito = (partidas_ganadas / partidas_jugadas) * 100
            
            ws[f"E{row}"] = partidas_jugadas
            ws[f"F{row}"] = partidas_ganadas
            ws[f"G{row}"] = partidas_perdidas
            ws[f"H{row}"] = intentos_totales
            ws[f"I{row}"] = promedio_intentos
            ws[f"J{row}"] = porcentaje_exito
            break

    # Si no se encuentra, añade nueva fila para el jugador, modo y  diferentes dificultades
    if not encontrado:
        partidas_jugadas = 1
        partidas_ganadas = 1 if Resultado else 0
        partidas_perdidas = 0 if Resultado else 1
        intentos_totales = Intentos
        promedio_intentos = Intentos
        porcentaje_exito = 100 if Resultado else 0
        nueva_fila = [Jugador, Modo, Dificultad1, Dificultad2, partidas_jugadas, partidas_ganadas, partidas_perdidas, 
                      intentos_totales, promedio_intentos, porcentaje_exito]
        ws.append(nueva_fila)
    
    wb.save('estad_tareaJF.xlsx')

# Muestra estadísticas en pantalla
def mostrar_estad():
    wb = load_workbook('estad_tareaJF.xlsx')
    ws = wb['Estadísticas']
    print(f"{'Jugador':<15}{'Modo':<10}{'Dif. 1':<10}{'Dif. 2':<10}{'Jugadas':<10}{'Ganadas':<10}{'Perdidas':<10}{'Intentos Tot.':<15}{'Promedio Int.':<15}{'Éxito (%)':<10}")
    for row in ws.iter_rows(min_row=2, values_only=True):
        print(f"{row[0]:<15}{row[1]:<10}{row[2]:<10}{row[3]:<10}{row[4]:<10}{row[5]:<10}{row[6]:<10}{row[7]:<15.2f}{row[8]:<15.2f}{row[9]:<10.2f}")

# Agrega estadísticas ficticias
def agregar_estad():
    fake = Faker()
    wb = load_workbook('estad_tareaJF.xlsx')
    print('Generando estadísticas. Espera un momento...')
    # Generar estadísticas ficticias
    for i in range(10):  # Crea 10 jugadores que juegan 50 partidas cada uno. En total 500 partidas.
        jugador = fake.first_name()
        for i in range(50):
            modo = random.choice(['Solitario', 'Pareja'])
            dificultad1 = random.choice(['Fácil', 'Normal', 'Difícil']) #Dificultad en función de los intentos
            dificultad2 = random.choice(['Fácil', 'Normal', 'Difícil']) #Dificultad en función del intervalo
            resultado = random.choice([True,False])
            if resultado == True : #Si gana calcula el número de intentos
                if dificultad1 == 'Fácil' :#Dependiendo de la dificultad en intentos calcula un número al alzar entre los intentos posibles.
                    intentos = random.randint(1,20)
                else:
                    if dificultad1 == 'Normal' :
                        intentos = random.randint(1,12)
                    else:
                        intentos = random.randint(1,5)
            else: #Si pierde los intentos serán los máximos dependiendo de la dificultad.
                if dificultad1 == 'Fácil' :
                    intentos = 20
                else:
                    if dificultad1 == 'Normal' :
                        intentos = 12
                    else:
                        intentos = 5
            #Actualiza el registro de excel con la función creada anteriormente actualizar_estad.
            #Si existe un registro del jugador en el mismo modo y dificultad lo actualiza y sino lo crea.
            actualizar_estad(jugador.upper(), modo, dificultad1, dificultad2, resultado, intentos)

#Filtra por jugador y muestra estadísticas
def jugador_estad():
    wb = load_workbook('estad_tareaJF.xlsx')
    ws = wb['Estadísticas']
    jugador = input('Introduce el nombre del jugador para ver las estadísticas: ').upper()
    contador = 0    #Al haber varias entradas en el fichero excel de un mismo jugador(guarda en función de dificultad) 
                    #necesito saber cuantas entrasdas por jugador hay
    #Variables para almacenar datos totales y así después sacar gráficos
    jugadas_tot = 0
    ganadas_tot = 0
    perdidas_tot = 0
    intentos_tot = 0
    promedio_intentos_tot = 0
    exito_tot = 0
    promedio_exito = 0
    
    #Itera sobra las filas y si encuentra que valor coincide con el nombre del jugador la imprime. Además actualiza los valores para un futuro gráfico.
    for row in ws.iter_rows(min_row=2, values_only=True):
        if row[0] == jugador:
            if contador == 0:
                 print(f"{'Jugador':<15}{'Modo':<10}{'Dif. 1':<10}{'Dif. 2':<10}{'Jugadas':<10}{'Ganadas':<10}{'Perdidas':<10}{'Intentos Tot.':<15}{'Promedio Int.':<15}{'Éxito (%)':<10}")
                
            print(f"{row[0]:<15}{row[1]:<10}{row[2]:<10}{row[3]:<10}{row[4]:<10}{row[5]:<10}{row[6]:<10}{row[7]:<15.2f}{row[8]:<15.2f}{row[9]:<10.2f}")

            jugadas_tot = jugadas_tot + row[4]
            ganadas_tot = ganadas_tot + row[5]
            perdidas_tot = perdidas_tot + row[6]
            intentos_tot = intentos_tot + row[8]
            exito_tot = exito_tot + row[9]
            contador += 1
            
    if contador == 0:
        print('Jugador no encontrado')

    else: # Al imprimir los datos del jugador, imprime despúes gráficos de barras mostrando un resumen de sus partidas.
        print()
        promedio_intentos_tot = intentos_tot / contador
        promedio_exito = exito_tot / contador
        promedio_perdidas = 100 - promedio_exito
        
        #Gráfico de barras
        eje_x = ['Jugadas','Ganadas','Perdidas','Media Intentos']
        eje_y = [ jugadas_tot, ganadas_tot, perdidas_tot, promedio_intentos_tot]
        grafico.bar( eje_x, eje_y, color = 'orange')

        #Pone los valores encima de cada barra
        for i, valor in enumerate(eje_y):
            grafico.text(i, valor + 0.5, f'{valor:.2f}' , ha='center', va='bottom')
            
        grafico.title(f'PARTIDAS TOTALES DE {jugador}')
        grafico.tight_layout()
        grafico.show()

        #Gráfico de pastel
        etiquetas = ['% Ganadas','% Perdidas']
        valores = [promedio_exito, promedio_perdidas]
        grafico.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=140)
        grafico.title(f'% DE EXITO DE {jugador}')
        grafico.tight_layout()
        grafico.show()
         
    
#Menú Estadísticas. Elige las estadísticas que quieres ver.
def Menu_Estad():
    while True:
        print("""      Menú de Estadísticas 
1. Ver estadísticas de todos los jugadores.
2. Ver estadísticas por jugador. (Incluye gráficos)
3. Agregar estadísticas ficticias.
4. Salir del menú de estadísticas.

Elige una opción ingresando el número correspondiente.""")
        
        opcion = gn.pedir_num(1,4)
        
        if opcion == 1:
            mostrar_estad()  # Muestra todas las estadísticas
            print()
        elif opcion == 2:
            jugador_estad() #Filtra estadísticas por jugador
            print()
        elif opcion == 3:
            agregar_estad()  # Genera datos ficticios de prueba
            print("Estadísticas ficticias agregadas.")
            print()
        else:
            print("Saliendo del menú de estadísticas.")
            break
            

#El menú principal del juego. Da la bienvenida, explica y pide una modalidad a elegir. Finalmente devuelve la modalidad elegida.
def Menu_Principal():
    print("""                                       ¡Bienvenido al juego de Adivina el Número!

¿Te gustan los retos? Este juego pone a prueba tu intuición y habilidad para adivinar números. Hay cuatro opciones disponibles:

1. Modo de Un Jugador:
   Aquí jugarás contra la computadora. Tendrás que adivinar un número aleatorio generado por el sistema en un número limitado de intentos.
   Cada pista te ayudará a acercarte más al número secreto.

2. Modo de Dos Jugadores:
   Un jugador introduce un número secreto dentro de un rango establecido, y el otro jugador tiene que adivinarlo.
   Recibirá pistas después de cada intento, indicándole si el número secreto es mayor o menor al que ha introducido.
   ¡Ideal para desafiar a un amigo!

3. Ver Estadísticas del Juego:
   Revisa tus estadísticas: la cantidad de partidas jugadas, el número de intentos promedio, recórds y el porcentaje de éxito según usuario.
   ¡Haz un seguimiento de tus progresos y mejora en cada juego!

4. Salir del Juego:
   Selecciona esta opción para finalizar el juego y cerrar el programa.

                      Selecciona tu modalidad, afina tu intuición, ¡y demuestra qué tan buen adivinador eres!
                                Elige una opción ingresando el número correspondiente.
                                
""")
    return gn.pedir_num(1,4)
    print()

#Para elegir la dificultad de los intentos y del intervalo a adivinar. 
#Guarda las variables de los intentos y los números entre los que se va a adivinar según la dificultad escogida.   
def Menu_Dificultad(): 
    global x
    global y
    global z
    global Dificultad1
    global Dificultad2

    #Empezamos seleccionando la dificultad en función de los intentos (Dificultad1)
    print('Selecciona la dificultad en función de los intentos: \n'
          ' 1. Fácil (20 intentos). \n 2. Medio (12 intentos). \n 3. Difícil (5 intentos). \n Elige una opción ingresando el número correspondiente.')
    Dificultad1 = gn.pedir_num(1,3)
    print()
    if Dificultad1 == 1:
        Dificultad1 = 'Fácil'
        z=20
    elif Dificultad1 == 2:
        Dificultad1 = 'Normal'
        z=12
    elif Dificultad1 == 3:
        Dificultad1 = 'Difícil'
        z=5   

    #(Opcional) Preguntamos si le gustaría añadir dificultad en función del intervalo. En caso afirmativo preguntamos por la dificultad requerida.
    # Por la contra, si elige No, se pondrá el intervalo de serie (1 a 1000), lo que equivale a una dificultad Normal.
    print('¿Te gustaría añadir dificultad en función de intervalo a adivinar? \n'
          ' 1. Sí. \n 2. No. (Default: de 1 a 1000) \n Elige una opción ingresando el número correspondiente.')
    q = gn.pedir_num(1,2)
    print()
    if q == 1:
        print('Selecciona la dificultad en función del intervalo: \n 1. Fácil (1 a 800). \n 2. Normal (1 a 1000). \n 3. Difícil (1 a 1200). \n '
              'Elige una opción ingresando el número correspondiente.')
        Dificultad2 = gn.pedir_num(1,3)
        print()
        if Dificultad2 == 1:
            Dificultad2 = 'Fácil'
            x = 1
            y = 800
        else:
            if Dificultad2 == 2: 
                Dificultad2 = 'Normal'
                x = 1
                y = 1000
            else:
                Dificultad2 = 'Difícil'
                x = 1
                y = 1200
            
    else:
        Dificultad2 = 'Normal'
        x = 1
        y = 1000

    #Informamos sobre las selecciones elegidas:
    print(f'Has elegido tener {z} intentos y adivinar entre los números {x} y {y}.', '\n')



#La función principal del juego. Tras escoger la dificultad y el número a elegir procede a desenvolver el juego. Comprueba que el jugador -
# - no introduzca un numero ya incluído.
def Juego(a,x,y,z,Jugador2,Modo,Dificultad1,Dificultad2):
    
    N=[]
    print(Jugador2,f': Adivina el número, tienes {z} intentos.', '\n')
    intentos=1
    
    while intentos<=z:
    
        print(f'INTENTO {intentos}') 
        n = gn.pedir_num(x,y)
        if n in N:
            print('Número ya introducido.')
            continue
            
        if n == a:
            print('\n FELICIDADES', Jugador2.upper(),'! LO HAS ADIVINADO. \n')
            actualizar_estad(Jugador2.upper(),Modo,Dificultad1,Dificultad2, Resultado = True, Intentos = intentos)
            break
        elif a > n:
            print('\n Has fallado. ES MAYOR. \n')
        else:
            print('\n Has fallado. ES MENOR. \n')

        N.append(n)
        intentos += 1
            
    else:
        print('\n HAS PERDIDO',Jugador2.upper()+'!','EL NÚMERO SECRETO ERA EL', str(a)+'.','PRUEBA DE NUEVO. \n')
        actualizar_estad(Jugador2.upper(),Modo,Dificultad1,Dificultad2, Resultado = False, Intentos = z)


#El menú final. Al final del juego o de las estadísticas pregunta si quiere volver a jugar o no. Devuelve True para jugar y False para salir.
def Menu_Final():
    print("""
            ----- MENÚ FINAL -----
            
            1. Volver a jugar
               (¿Listo para otro desafío?)
               
            2. Salir del juego
               (Para finalizar tu experiencia)
               
    Elige una opción ingresando el número correspondiente.
    
""")
    
    f = gn.pedir_num(1,2)
    print()
    if f==1:
        return True

    else:
        return False


#El mensaje final. Creo uno función para no tener que ponerlo varias veces en el código y que este quede más simple y ordenado.
def Mensaje_Final():
     print("""
            ¡Gracias por jugar a Adivina el Número!
            
    Esperamos que te hayas divertido y que vuelvas pronto 
     para seguir mejorando tus habilidades de adivinanza.
          
                       ¡Hasta la próxima!
                       
""")

#Comienza el juego

m = None
while m != 4 :
    m = Menu_Principal()

    if m==1 or m==2 :
        abrir_estad()
        Menu_Dificultad()

        if m==1 : #Modo Solitario
            Modo = 'Solitario'
            J1 = input('Introduce tu nombre: ')
            print()
            a = random.randint(x,y)
            Juego(a,x,y,z,J1,Modo,Dificultad1,Dificultad2)
            
            if not(Menu_Final()):
                Mensaje_Final()
                break
        else: #Modo en Pareja
            Modo = 'Pareja'
            J1 = input('Jugador 1, introduce tu nombre: ')
            J2 = input('Jugador 2, introduce tu nombre: ')
            print()
            a = gn.secret_num(x,y,J1)
            Juego(a,x,y,z,J2,Modo,Dificultad1,Dificultad2)

            if not(Menu_Final()):
                    Mensaje_Final()
                    break

    elif m==3:
        abrir_estad()
        Menu_Estad()
        if not(Menu_Final()):
                    Mensaje_Final()
                    break

    else:
        Mensaje_Final()
        


            
```
